# **Instrodução a processamento de linguagem natural**

#### O que é NLP
Processamento de linguagem natural é uma área dentro da Inteligência Artificial, dedicada a desenvolver a capacidade de máquinas compreender e até escrever a linguagem natural dos seres humanos.

Este tipo de tecnologia está presente em diversos dispositivos presentes em nosso dia a dia, como Smartphones, SmartWatches, SmartTVS, Assistentes digitais, entre outros.


**NLP se divide entre duas partes principais.**

NLP (natural language processing) se refere ao mundo de NLP de modo geral e genérico e se divide em duas frentes:

- NLU (natural language understanding) trata do entendimento, compreensão da linguagem natural, está associada a tarefas como análise de sentimentos, busca, classificação de textos e sumarização de textos.

- NLG (natural language generation) foca na criação de linguagem natural, ou seja, textos. A aplicação disto pode ser desde respostas a interações com Bots a textos complexos e completos.

#### **NLP Técnicas**

1. **Data Cleaning e pré-processamento**

A limpeza dos dados é necessária em qualquer projeto de IA ou DS, em NLP nos preocupamos em padronizar os textos, extrair sujeiras como html, emojis, pontuações pois muitos textos podem vir de crawler de sites, e com isto vem coisas que não são úteis para nossas análises e modelos. 

Um dos primeiros pontos de atenção é com a retirada de stopwords. As stopwords são “palavras com pouca informação”. São palavras de conexão, “o”, “de”, “que”, ou palavras muito comuns como “para”. 

Para realizar estas tarefas temos bibliotecas com modelos pré-treinados em diferentes idiomas que ajudam a extrair as stopwords e outras técnicas. 


2. **Stemming e Lemmatization**

Stemming e lemmatization são processos focados em reduzir as palavras a sua raiz. Por exemplo “Correu”, “Corrido”, “Correndo” são todos derivados da mesma palavra “correr” e geralmente possuem o mesmo significado, passando a informação, desconsiderando o tempo gramatical. 

Essas transformações é uma boa forma de reduzir a dimensionalidade e ruídos no texto. A diferença entre Stemming e Lemmatization é que Lemmatization identifica a correspondência entre correr e correu, bem como a correspondência entre correr e correria. Já Stemming para na identificação de correr e correu. 

Para idiomas morfologicamente mais complexos, como é a língua portuguesa, é indicado Lemmatization para que possa ser identificado mais facilmente as correspondências entre as palavras.


3. **Tokenization**

A tokenização é o processo que separa as palavras ou frases umas das outras, isso ajuda em muitos cenários como na contagem de palavras e/ou frases. A tokenização é um processo necessário para a aplicação das técnicas de NLP.

As bibliotecas NLTK e Spacy possuem todo tipo de tratativa para processamento de linguagem natural.

4. **TF-IDF Term Frequency-inverse document frequency**

O TF-IDF é uma técnica que vetoriza as palavras com base em sua importância que elas representam. Se uma palavra aparece com frequência em um documento, esse termo provavelmente tem importância neste documento. Mas se essa mesma palavra aparece com frequência em muitos documentos então é um termo de menor importância pensando em um escopo global. 

O TF-IDF faz uma estimativa do trade-off entre as duas abordagens, atribuindo números aos tokens com base na frequência com que cada token aparece. 

5. **BOW - Bag of Words**

O algoritmo de bag-of-words é muito conhecido e já foi amplamente discutido e implementado por outros desenvolvedores. O modelo bag-of-words é uma representação simplificada usada no processamento de linguagem natural. Nesse modelo, um texto (como uma frase ou um documento) é representado como a bolsa de suas palavras, desconsiderando a gramática e até mesmo a ordem das palavras, mas mantendo a multiplicidade. 
Exemplo:

6. **POS Tagging**

O POS Tagging se trata de um processo de ro-tulação de elementos textuais-tipicamente palavras e pontuação-com o fim de evidenciar a estrutura gramatical de um determinado trecho de texto.
Em outras palavras, pos tagging consegue determinar se uma palavra é um subjstantivo, pronome, verbo, adverbio, etc.


7. **Análise de dependencia**

A análise de dependencia apresenta a relação entre as palavras em um texto. Uma mesma palavra ´pode assumir diferentes papeis em um texto, hora como substativo, hora como verbo para executar uma ação. A análise de dependencia vai criar essa relação para que possam ser levado em conta por modelos matemáticos na hora de criar um modelo de NLP.

8. **NER (Named Entity Recognition)**

Named Entity Recognition (NER), em português Reconhecimento de Entidade Nomeada, é uma técnica de Natural Processing Language (NLP) que consiste em identificar e categorizar informações-chave (entidades) em textos. Uma entidade pode ser qualquer palavra ou série de palavras que se referem ao mesmo tema. Cada entidade detectada é classificada em uma categoria predeterminada. Por exemplo, as entidades podem ser nomes de pessoas, organizações, locais, horários, quantidades, valores monetários, porcentagens, entre outros.

Diversas bibliotecas possuem modelos treinados para detecção de entidades nomeadas, mas é possível treinar modelos com entidades personalizadas.

```python
# Limpando tags de html 
import spacy 

nlp = spacy.load("en_core_web_sm")

import warnings 

warnings.simplefilter("ignore")
from bs4 import BeautifulSoup 
import requests
import traceback

def get_text (url):
    buf = []
    
    try:
        soup = BeautifulSoup(requests.get(url).text, "html.parser")
        
        for p in soup.find_all("p"):
            buf.append(p.get_text())

        return "\n".join(buf)
    except:
        print(traceback.format_exc())
        sys.exit(-1)
```

```python
lic = {}
lic["mit"] = nlp(get_text("https://opensource.org/licenses/MIT"))
lic["asl"] = nlp(get_text("https://opensource.org/licenses/Apache-2.0"))
lic["bsd"] = nlp(get_text("https://opensource.org/licenses/BSD-3-Clause"))

for sent in lic["bsd"].sents:
    print(">", sent)
```

```python
for sent in lic["asl"].sents:
    print(">", sent)
```

```python
for sent in lic["bsd"].sents:
    print(">", sent)
```

```python
# Comparação entre os textos por similaridade

pairs = [
         ["mit", "asl"],
         ["asl", "bsd"],
         ["bsd", "mit"]
]
for a, b in pairs:
  print(a,b,lic[a].similarity(lic[b]))
```

```python
domains = ["finance", "banking"]
sentence = nlp("I want to withdraw 5,000 euros.")

enriched_sent = []

for token in sentence:
    synsets = token._.wordnet.wordnet_synsets_for_domain(domains)
    
    if synsets:
        lemmas_for_synset = []
        
        for s in synsets:
            lemmas_for_synset.extend(s.lemma_names())
            enriched_sent.append("({})".format("|".join(set(lemmas_for_synset))))
    else:
        enriched_sent.append(token.text)

print(" ".join(enriched_sent))
```
